### Maintenance Prediction as a classification problem

```python
df_train
```

```python
cycle=30
df_train['label'] = df_train['RUL'].apply(lambda x: 1 if x <= cycle else 0)
```

```python
op_set=["op"+str(i) for i in range(1,4)]
sensor=["sensor"+str(i) for i in range(1,22)]
```

```python
df_train.label.unique()
```

```python
df_train['label'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X=df_train.drop(["id","cycle","op3","sensor1","sensor5","sensor6","sensor10","sensor16","sensor18","sensor19","RUL",'label'],axis=1)
y=df_train['label']
```

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)
X, y = ros.fit_resample(X, y)

# Create a new dataframe with resampled data
df_train= pd.concat([pd.DataFrame(X), pd.Series(y, name='label')], axis=1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,y, test_size = 0.20, random_state = 42)
```

```python
df_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(X_train,Y_train)
train_pred = lr.predict(X_train)
test_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(Y_test,test_pred)
```

```python
accuracy_score(Y_test,test_pred)
```

```python
confusion_matrix(Y_train,train_pred)
```

```python
accuracy_score(Y_train,train_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,Y_train)
dtc_train_pred = dtc.predict(X_train)
dtc_test_pred = dtc.predict(X_test)
```

```python
confusion_matrix(Y_train,dtc_train_pred)
```

```python
accuracy_score(Y_train,dtc_train_pred)
```

```python
confusion_matrix(Y_test,dtc_test_pred)
```

```python
accuracy_score(Y_test,dtc_test_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,Y_train)
rfc_train_pred = rfc.predict(X_train)
rfc_test_pred = rfc.predict(X_test)
```

```python
confusion_matrix(Y_train,rfc_train_pred)
```

```python
accuracy_score(Y_train,rfc_train_pred)
```

```python
confusion_matrix(Y_test,rfc_test_pred)
```

```python
accuracy_score(Y_test,rfc_test_pred)
```

from doing the classification models like linear regression, decision tree and random forest we can conlcude that for logistic regression model both training and testing data show similar accuarcy and we choose that mode
